```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the uploaded files
pune_during_df = pd.read_excel('C:/Users/karan/Downloads/1/2. Data cleaned/Pune_During_Prepared.xlsx')
pune_post_df = pd.read_excel('C:/Users/karan/Downloads/1/2. Data cleaned/Pune_Post_Prepared.xlsx')

# Print the columns to debug
print(pune_during_df.columns)
print(pune_post_df.columns)

# Define the KPI columns for scaling (adjust based on actual columns)
kpi_columns = ['Impressions', 'Reach', 'Likes', 'Love Reactions', 'Care Reactions', 
               'Haha Reactions', 'Wow Reactions', 'Sad Reactions', 'Angry Reactions', 
               'Comment Count', 'Comment Replies', 'Share Count', 'Shared Reach', 
               'Link Clicks', 'Profile Clicks', 'Video Views', 'Hide Post', 
               'Report as Spam', 'Unfollow Page', 'CTR (%)', 'Conversion Rate (%)']

# Select the KPI columns for both datasets
pune_during_kpi = pune_during_df[kpi_columns]
pune_post_kpi = pune_post_df[kpi_columns]

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler on the Pune during dataset and transform both datasets accordingly
pune_during_scaled = scaler.fit_transform(pune_during_kpi)
pune_post_scaled = scaler.transform(pune_post_kpi)

# Convert the scaled data back into DataFrames for easy handling
pune_during_scaled_df = pd.DataFrame(pune_during_scaled, columns=kpi_columns)
pune_post_scaled_df = pd.DataFrame(pune_post_scaled, columns=kpi_columns)

# Save the scaled data to Excel files
pune_during_scaled_df.to_excel('C:/Users/karan/Downloads/1/2. Data cleaned/Scaled_Pune_During_Data.xlsx', index=False)
pune_post_scaled_df.to_excel('C:/Users/karan/Downloads/1/2. Data cleaned/Scaled_Pune_Post_Data.xlsx', index=False)

# Provide paths to the user
output_files_paths = [
    'C:/Users/karan/Downloads/1/2. Data cleaned/Scaled_Pune_During_Data.xlsx',
    'C:/Users/karan/Downloads/1/2. Data cleaned/Scaled_Pune_Post_Data.xlsx'
]

output_files_paths
```
