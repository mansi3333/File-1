```python

import pandas as pd

# Load the data from the uploaded Excel files
nashik_data = pd.read_excel(r'C:\Users\karan\Downloads\1\Nashik_PreCampaignData.xlsx')
nagpur_data = pd.read_excel(r'C:\Users\karan\Downloads\1\Nagpur_PreCampaignData.xlsx')
thane_data = pd.read_excel(r'C:\Users\karan\Downloads\1\Thane_PreCampaignData.xlsx')
navi_mumbai_data = pd.read_excel(r'C:\Users\karan\Downloads\1\NaviMumbai_PreCampaignData.xlsx')
pune_post_data = pd.read_excel(r'C:\Users\karan\Downloads\1\Pune_PostCampaignData.xlsx')
pune_during_data = pd.read_excel(r'C:\Users\karan\Downloads\1\Pune_DuringCampaignData.xlsx')
pune_pre_data = pd.read_excel(r'C:\Users\karan\Downloads\1\Pune_PreCampaignData.xlsx')  # Added this line

# Display the first few rows of each dataset to understand their structure
data_samples = {
    "Nashik Pre-Campaign": nashik_data.head(),
    "Nagpur Pre-Campaign": nagpur_data.head(),
    "Thane Pre-Campaign": thane_data.head(),
    "Navi Mumbai Pre-Campaign": navi_mumbai_data.head(),
    "Pune Post-Campaign": pune_post_data.head(),
    "Pune During-Campaign": pune_during_data.head(),
    "Pune Pre-Campaign": pune_pre_data.head()  # Added this line
}

data_samples
```

```python
# Function to prepare data
def prepare_data(df):
    # Checking for missing values and filling them (using forward fill as an example)
    df.fillna(method='ffill', inplace=True)
    
    # Removing duplicates
    df.drop_duplicates(inplace=True)
    
    # Ensuring date format consistency
    df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
    
    return df

# Prepare the Nashik data
nashik_prepared = prepare_data(nashik_data.copy())

# Prepare the Pune Pre-Campaign data (added this step)
pune_pre_prepared = prepare_data(pune_pre_data.copy())

# Show the first few rows of the cleaned Nashik and Pune Pre-Campaign data
nashik_prepared.head(), nashik_prepared.info(), pune_pre_prepared.head(), pune_pre_prepared.info()
```

```python
# Apply the preparation function to the remaining datasets
nagpur_prepared = prepare_data(nagpur_data.copy())
thane_prepared = prepare_data(thane_data.copy())
navi_mumbai_prepared = prepare_data(navi_mumbai_data.copy())
pune_post_prepared = prepare_data(pune_post_data.copy())
pune_during_prepared = prepare_data(pune_during_data.copy())

# Showing the first few rows of the cleaned data for each city to ensure the preparation was successful
prepared_data_samples = {
    "Nagpur Pre-Campaign": nagpur_prepared.head(),
    "Thane Pre-Campaign": thane_prepared.head(),
    "Navi Mumbai Pre-Campaign": navi_mumbai_prepared.head(),
    "Pune Post-Campaign": pune_post_prepared.head(),
    "Pune During-Campaign": pune_during_prepared.head()
}

# Check for information summary to ensure no null values and proper data types
info_data = {
    "Nagpur Info": nagpur_prepared.info(),
    "Thane Info": thane_prepared.info(),
    "Navi Mumbai Info": navi_mumbai_prepared.info(),
    "Pune Post Info": pune_post_prepared.info(),
    "Pune During Info": pune_during_prepared.info()
}

prepared_data_samples, info_data
```
